**Exercise 2**

```python
# Constants
PREMIUM = 1.5
NORMAL_HOURS = 40

def get_pay(hours: float, rate: float) -> float:
    """Compute gross pay, accounting for overtime."""
    if hours > NORMAL_HOURS:
        hours = (hours - NORMAL_HOURS)*PREMIUM + NORMAL_HOURS
    return hours * rate

if __name__ == '__main__':
    # Get user input
    hours = float(input('Enter Hours: '))
    rate = float(input('Enter hourly rate: '))

    # Calculate gross pay
    pay = get_pay(hours, rate)

    # Print the result
    print(f'Pay: {pay:.1f}')
```

```python
try:
    # Prompt user for input
    score = float(input("Enter the score (between 0 and 100): "))

    # Check if the input score is within the valid range
    if 0 <= score <= 100:
        # Determine the grade based on the score
        if 97 <= score <= 100:
            grade = 1.0
        elif 93 <= score < 97:
            grade = 1.25
        elif 89 <= score < 93:
            grade = 1.5
        elif 85 <= score < 89:
            grade = 1.75
        elif 81 <= score < 85:
            grade = 2.0
        elif 77 <= score < 81:
            grade = 2.25
        elif 74 <= score < 77:
            grade = 2.5
        elif 0 <= score < 74:
            grade = 3.0
        # Print the calculated grade
        print("Grade:", grade)
    else:
        # If the input score is outside the valid range, print an error message
        print("Score must be between 0 and 100.")
except ValueError:
    # If the input is not a valid number, print an error message
    print("Invalid input. Please enter a number.")
```
