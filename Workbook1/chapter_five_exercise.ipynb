**Exercise 1**

```python
a = 2
b = 2.0
c = a

# comparing values using ==
print("Using ==:")
print("a == b:", a == b)  # true because the values are equal
print("a == c:", a == c)  # and also true because values are equal
print()

# comparing identities using is
print("Using is:")
print("a is b:", a is b)  # false because they are different objects
print("a is c:", a is c)  # true because they refer to the same object
```

Challenge 1

```python
total = 0
count = 0

while True:  # Using an infinite loop
    num_input = input("Enter a number: ")

    if num_input == 'done':  # If the user enters 'done', it will end the loop
        break

    try:  # using the try-except block
        num = float(num_input)
        total += num
        count += 1
    except ValueError:
        print("Invalid input")
        continue

if count > 0:  # calculating the average using if-else
    average = total / count
    print(total, count, average)
else:
    print("No numbers entered")
```

**Challenge 2**


```python
numbers = []  # initialize an empty list to store numbers

while True:
    num_input = input("Enter a number: ")

    if num_input == 'done':
        break

    try:
        num = float(num_input)
        numbers.append(num)  # add the number to the list
    except ValueError:
        print("Invalid input")
        continue

if numbers:  # check if the list is not empty
    maximum = max(numbers)  # will find the maximum number
    minimum = min(numbers)  # will find the minimum number
    print("Maximum:", maximum)
    print("Minimum:", minimum)
else:
    print("No numbers entered")
```
