```python
def compute_pay(hours, rate):  # Function to calculate pay based on hours and rate
    if hours <= 40:  # If hours worked is 40 hours or less
        pay = hours * rate  # Calculate pay for regular hours
    else:  # If hours worked is more than 40 hours will execute the else
        regular_hours = 40  # Set regular hours to 40
        overtime_hours = hours - regular_hours  # Calculate overtime hours
        regular_pay = regular_hours * rate  # Calculate pay for regular hours
        overtime_pay = overtime_hours * (rate * 1.5)  # Calculate pay for overtime hours with time-and-a-half
        pay = regular_pay + overtime_pay  # Total pay is sum of regular and overtime pay
    return pay

hours = float(input("Enter Hours: "))  # Prompt user to input hours worked
rate = float(input("Enter Rate: "))  # Prompt user to input hourly rate

pay = compute_pay(hours, rate)
print("Pay:", pay)
```

```python
def computegrade(score):
     # Assigning grades based on score ranges
    if score >= 97: # Highest grade
        grade = '1.0'
    elif score >= 93:
        grade = '1.25'
    elif score >= 89:
        grade = '1.5'
    elif score >= 85:
        grade = '1.75'
    elif score >= 81:
        grade = '2.0'
    elif score >= 77:
        grade = '2.25'
    elif score >= 74:
        grade = '2.5'
    else:
        grade = '3.0'  # Lowest grade
    return grade

# Test the function with different score values
scores = [100, 97, 93, 89, 85, 81, 77, 74, 50]

for score in scores:
    print(f"Score: {score}  Grade: {computegrade(score)}")
```
